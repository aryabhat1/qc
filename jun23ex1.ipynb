```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
```

```python
qc_cnot = QuantumCircuit(2)
qc_cnot.cx(0, 1)
qc_cnot.draw('mpl')
```

```python
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0, 1)
qc.measure(0, 0)
qc.measure(1, 1)
qc.draw('mpl')
```

```python
qc_ha = QuantumCircuit(4, 2)
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()
qc_ha.cx(0, 2)
qc_ha.cx(1, 2)
qc_ha.barrier()
```

```python
qc_ha.measure(2, 0)
qc_ha.measure(3, 1)
# qc_ha.measure(3, 0)
```

```python
qc_ha.draw('mpl')
```

```python
qc_ha = QuantumCircuit(4, 2)
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()
qc_ha.cx(0, 2)
qc_ha.cx(1, 2)
qc_ha.ccx(0, 1, 3)
qc_ha.barrier()
qc_ha.measure(2, 0)
qc_ha.measure(3, 1)
qc_ha.draw('mpl')
```

```python
counts = execute(qc_ha, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
